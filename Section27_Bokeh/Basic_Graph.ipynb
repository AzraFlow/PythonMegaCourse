```python
# Making a basic Bokeh line graph

# importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

# Prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# Prepare the output file
output_file('Line.html')

# Create a figure object
f = figure()

# Create line plot
f.line(x, y)

# Write the plot in the figure object
show(f)
```

```python
# Making a second Bokeh graph with triangle glyphs

# Prepare the data
x = [3, 7.5, 10]
y = [3, 6, 9]

#Prepare the output file
output_file('triangle_glyphs.html')

# Create figure object
f = figure()

# Create point plot
f.triangle(x, y, size=10)

# Write the plot in the figure object
show(f)
```

```python
# Making a third Bokeh graph with circle glyphs

# Prepare the data
x = [3, 7.5, 10]
y = [3, 6, 9]

#Prepare the output file
output_file('circle_glyphs.html')

# Create figure object
f = figure()

# Create point plot
f.circle(x, y, size=10)

# Write the plot in the figure object
show(f)
```

```python
# Making a x, y line plot using Bokeh

# importing Bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

# Prepare some data
df = pandas.read_csv('data.csv')
x=df['x']
y=df['y']

# Prepare the output file
output_file('Line_from_csv.html')

# Create a figure object
f = figure()

# Create line plot
f.line(x, y)

# Write the plot in the figure object
show(f)
```

```python
# Plot Properties

# Import Bokeh and pandas
import pandas
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=500, plot_height=400, tools='pan', background=None)

p.title.text='Cool Data'
p.title.text_color='Gray'
p.title.text_font='times'
p.title.text_font_style='bold'
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label='Date'
p.yaxis.axis_label='intensity'

p.line([1, 2, 3], [4, 5, 6])
output_file('graph.html')
show(p)
```

```python
# Plot Weather Data

# Import Bokeh and pandas
import pandas
from bokeh.plotting import figure, output_file, show

plot = figure(tools='pan', plot_width=500, plot_height=400)

plot.title.text='Temperature and Air Pressure'
plot.title.text_color='Gray'
plot.title.text_font='arial'
plot.title.text_font_style='bold'
plot.xaxis.minor_tick_line_color=None
plot.yaxis.minor_tick_line_color=None
plot.xaxis.axis_label='Temperature (Â°C)'
plot.yaxis.axis_label='Pressure (hPa)'

# Prepare some data
df = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx")
x=df['Temperature']/10
y=df['Pressure']/10

# Create scatter plot
plot.circle(x, y, size=0.5, color='blue', alpha=0.5)

# Prepare the output file
output_file('Weather_Data_Plot.html')

# Write the plot in the figure object
show(plot)
```

```python
# Visual Attribute Example

# Import Bokeh and pandas
import pandas
from bokeh.plotting import figure, output_file, show

plot = figure(plot_width=500, plot_height=400, tools='pan, reset')

plot.title.text='Earthquakes'
plot.title.text_color='Orange'
plot.title.text_font='times'
plot.title.text_font_style='italic'
plot.xaxis.minor_tick_line_color='Yellow'
plot.yaxis.minor_tick_line_color='Yellow'
plot.xaxis.axis_label='Times'
plot.yaxis.axis_label='Value'

# Create plot
plot.circle([1, 2, 3, 4, 5], [5, 6, 5, 5, 3], size=[i*2 for i in [8, 12, 14, 15, 20]], color='red', alpha=0.5)

# Prepare the output file
output_file('Scatter_Plot.html')

# Write the plot in the figure object
show(plot)
```

```python
# Import Bokeh and pandas
import pandas
from bokeh.plotting import figure, output_file, show

plot = figure(plot_width=500, plot_height=250, x_axis_type='datetime', sizing_mode='scale_height')

plot.title.text='Timeseries'
plot.title.text_color='Orange'
plot.title.text_font='times'
plot.title.text_font_style='italic'
plot.xaxis.minor_tick_line_color='Yellow'
plot.yaxis.minor_tick_line_color='Yellow'
plot.xaxis.axis_label='Times'
plot.yaxis.axis_label='Value'

# Prepare some data
df = pandas.read_csv("adbe.csv", parse_dates=["Date"])

# Create plot
plot.line(df["Date"], df["Close"], color='orange', alpha=0.5)

# Prepare the output file
output_file('Timeseries.html')

# Write the plot in the figure object
show(plot)
```

```python
# Import Bokeh and pandas
import pandas
from bokeh.plotting import figure, output_file, show

p = figure(plot_width=500, plot_height=400, tools='pan, reset')

p.title.text = "Earthquake"
p.title.text_color = 'Orange'
p.title.text_font = 'times'
p.title.text_font_style = 'italic'

p.yaxis.minor_tick_line_color = 'Yellow'

p.xaxis.axis_label = 'Times'
p.yaxis.axis_label = 'Value'

p.line([1, 2, 3, 4, 5], [5, 6, 5, 5, 3], line_width=2, color='red', alpha=0.5)
p.circle([i*2 for i in [1, 2, 3, 4, 5]], [5, 6, 5, 5, 3], size=8, color='olive', alpha=0.5)

output_file('Line_with_Scatter_Plot.html')
show(p)
```
